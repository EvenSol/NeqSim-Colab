# Getting started as a NeqSim Java developer
The NeqSim source code is downloaded by cloning the library to your local computer (alternatively fork it to your private reprository). The following commands are dependent on a local installation of GIT and Maven.

In this demonstration we illustrate how to get started as a developer in three steps:

1. Cloning NeqSim to the local disk
2. Build NeqSim using Maven
3. Make changes to the code and update code base using Git

# 1. Clone neqsim to you local disk using GIT
NeqSim is cloned to you local disk using the git clone command. Alternatively clone neqsim into you own GitHub user, and clone neqsim from your own user. 

```
!git clone https://github.com/equinor/neqsim.git
```

# 2. Build NeqSim using Maven
NeqSim is built using the Maven wrapper command mvnw install. The NeqSim source code in the src/main directory will be built and the tests in the src/test directory will run. THe compiled NeqSIm library is found in the target directory (NeqSim.jar with all dependent libraries). 

```
%cd neqsim
!./mvnw install
```

# 3. Modify/add code and update neqsim via git
1. Modify or add a file, and recompile mvn install.
2. Commit and push the modifications to the central neqsim reprository using git commit and git push 

```
!git branch testing
!git checkout testing
!git status
!git commit *
# !git push --set-upstream origin testing   # pushing to the remote repository requires username/password correctly
```
